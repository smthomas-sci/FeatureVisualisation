```python
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt
import os

from skimage.transform import resize
```

```python
img_dir = "/home/simon/Desktop/10_tabby"
eagles = os.listdir(img_dir)
```

```python
stack = np.zeros((224, 224, 3))
```

```python
for file in eagles:
    img = imread(os.path.join(img_dir, file))[:, :, 0:3]
    
    img = resize(img, (224, 224))
    
    stack += img
```

```python
avg_img = stack / len(eagles)
```

```python
plt.imshow(avg_img)
plt.show()
```

```python
imsave("/home/simon/Desktop/avg_tabby.jpg", avg_img)
```
